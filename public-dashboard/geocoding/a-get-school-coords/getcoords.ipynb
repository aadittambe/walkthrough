```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import re
import math
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv("../../data/processed/etl_school_list.csv",dtype = str).replace({'&#45':'-', r'&#39':'', '&amp':'and'}, regex=True).replace(r'\n',' ', regex=True) 
# data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('.', '_')
data
```

```python
data["lat"] = ""
data["lon"] = ""
```

```python
counter = 0
for index,row in data.iterrows():
    # numpy.isnan(number)
    address = f"{row['street']}, {row['city']}, {row['state']} {row['zip']}"
    # print(address)
    geolocator = Nominatim(user_agent="Aadit Tambe, University of Maryland")
    try:
        counter = counter + 1
        location = geolocator.geocode(address)
        data["lat"][index] = location.latitude
        data["lon"][index] = location.longitude
        print(str(counter) + ': done')
        # time.sleep(2)
    except:
        counter = counter + 1
        data["lat"][index] = "no_result"
        data["lon"][index] = "no_result"
        print(str(counter) +': no result')
        # time.sleep(2)
    # data.to_csv("../../data/processed/etl_school_full_data_with_coords.csv", index=False)
```

```python
data.to_csv("../../data/processed/school_list_geocoded.csv", index=False)
```
